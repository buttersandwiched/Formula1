```python
from pyspark.sql.types import StringType, IntegerType, DoubleType, StructType, StructField
from pyspark.sql.functions import col, concat,current_timestamp, lit
```

```python
pitstops_schema = StructType([StructField("driverId", IntegerType(), True),
                              StructField("duration", StringType(), True),
                              StructField("lap", IntegerType(), True),
                              StructField("milliseconds", IntegerType(), True),
                              StructField("raceId", IntegerType(), True),
                              StructField("stop", IntegerType(), True)
                            ])
```

```python
pitstops_df = spark.read \
    .option("multiline", True) \
    .schema(pitstops_schema) \
    .json("abfss://dropzone@databrickssandbox202307.dfs.core.windows.net/formula1/pit_stops.json")
```

```python
pitstops_formatted_df = pitstops_df \
    .withColumn("ingestion_date", current_timestamp()) 
```

```python
pitstops_formatted_df.write.mode("overwrite").parquet("abfss://datalakehouse@databrickssandbox202307.dfs.core.windows.net/schemas/formula1/raw/pit_stops")
```
