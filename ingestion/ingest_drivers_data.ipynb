```python
from pyspark.sql.types import StringType, IntegerType, DoubleType, StructType, StructField
from pyspark.sql.functions import col, concat,current_timestamp, lit
```

```python
drivers_schema = StructType( \
    [StructField("code", StringType(), False),
     StructField("dob", StringType(), True),
     StructField("driverId", IntegerType(), False),
     StructField("driverRef", StringType(), False),
     StructField("name", StructType( \
         [ \
             StructField("forename", StringType(), True), 
             StructField("surname", StringType(), True)
          ]), True),
     StructField("nationality", StringType(), True),
     StructField("number", StringType(), True),
     StructField("url", StringType(), True)
     ])
```

```python
drivers_df = spark.read \
    .schema(drivers_schema) \
    .json("abfss://dropzone@databrickssandbox202307.dfs.core.windows.net/formula1/drivers.json")
```

```python
drivers_formatted_df = drivers_df \
    .withColumn("ingestion_timestamp", current_timestamp())
```

```python
drivers_formatted_df.write.mode("overwrite").parquet("abfss://datalakehouse@databrickssandbox202307.dfs.core.windows.net/schemas/formula1/raw/drivers")
```
