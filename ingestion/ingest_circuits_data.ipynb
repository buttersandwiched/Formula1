```python
from pyspark.sql.types import StringType, IntegerType, DoubleType, StructType, StructField
from pyspark.sql.functions import col, current_timestamp
```

```python
circuits_schema = StructType([StructField("circuitid", IntegerType(), True),
                              StructField("circuitref", StringType(), True),
                              StructField("name", StringType(), True),
                              StructField("location", StringType(), True),
                              StructField("country", StringType(), True),
                              StructField("lat", DoubleType(), True),
                              StructField("lng", DoubleType(), True),
                              StructField("alt", IntegerType(), True),
                              StructField("url", StringType(), True)
])
```

```python
circuits_df = spark.read \
    .option("header", True) \
    .schema(circuits_schema) \
    .csv(f"{dropzone_container}/circuits.csv")
```

```python
circuits_df_selected = circuits_df.withColumn(
    "date_ingested", current_timestamp()
    )
```

```python
circuits_df_selected.write \
    .mode("overwrite") \
    .parquet(f"{raw_container}/circuits")
```
